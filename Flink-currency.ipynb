```python
# import the required library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
fifth_column = []
sixth_column = []
seventh_column = []
count_list = []
counter = 0
```

```python
file_df = pd.read_csv('/home/nabil/eclipse-workspace/MovieRatingFlink/src/main/resources/latency/ccurrencyconversionTest.csv', header=None)
```

```python
frames = [file_df]
result = pd.concat(frames)

```

```python
result[4].iloc[-1]-result[2].iloc[0]
```

```python
result[2].iloc[0]
```

```python
for index, row in result.iterrows():
        fifth_column.append(int(row[4]) - int(row[2]))
        sixth_column.append(int(row[4]) - int(row[3]))
        seventh_column.append(int(row[4]) - result[2].iloc[0])
        count_list.append(counter)
        counter = counter + 1
```

```python
result[5] = fifth_column
result[6] = sixth_column
result[7] = seventh_column
```

```python
print('Average value of processing column: {}'.format(result[6].mean()))
print('Average value of event column: {}'.format(result[5].mean()))
```

```python
result.to_csv('frontierHighbid.csv', header=None)
```

```python
df = pd.DataFrame({
    'rows' : count_list,
   'processing': result[6],
   'event': result[5],
    'timex': result[4],
    'epo' : result[7]
   })
```

```python
df.plot(kind ="line", xlim=(0,100000),ylim=(0,17), y=["processing"],figsize=(10,5), grid=True, color= "orange")
```
